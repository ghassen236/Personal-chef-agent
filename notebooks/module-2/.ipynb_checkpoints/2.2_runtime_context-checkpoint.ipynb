So far we've learned to give our agents access to a lot of information: prompts, Chat memory, web queries, MCP resources.   

That's not all you can inform your agent with! What if we had some helpful context ?  The agent should know when it runs.


For example, the user it's currently talking to is an internal employee, or their preferred language is Spanish. 

We can create a context schema to fill in and pass to the agent, which the agent can then draw from to inform its actions. 

```python
from dotenv import load_dotenv

load_dotenv()
```

It loads your secret API keys from a .env file into your Python environment.

Imagine your .env file is a locked safe with your passwords:

And your Python code needs those keys:

load_dotenv() unlocks the safe and makes the keys available to your code.




```python
from dataclasses import dataclass

@dataclass
class ColourContext:
    favourite_colour: str = "blue"
    least_favourite_colour: str = "yellow"
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI
import os

# Get your key from environment
google_key = os.getenv("GOOGLE_API_KEY")
print(f"Key exists: {bool(google_key)}")

# Create the model
model = ChatGoogleGenerativeAI(
    model="models/gemini-2.5-flash",  # Use flash for free tier
    google_api_key=google_key
)
```

```python
from langchain.agents import create_agent

agent = create_agent(
    #model="gpt-5-nano",
    model= model,
    context_schema=ColourContext  
)
```

```python
from langchain.messages import HumanMessage

response = agent.invoke(
    {"messages": [HumanMessage(content="What is my favourite colour?")]},
    context=ColourContext()
)
```

```python
from pprint import pprint

pprint(response)
```

```python
def clean_response(response):
    """Extract clean text from agent response"""
    last_msg = response['messages'][-1]
    
    if isinstance(last_msg.content, list):
        # Handle Gemini's list format
        for item in last_msg.content:
            if isinstance(item, dict) and item.get('type') == 'text':
                return item['text']
    else:
        return last_msg.content
    
    return "No text content found"

```

```python
print(clean_response(response))
```

The context isn't actually passed to the model directly. Instead it's passed to the tool cause in an object called tool runtime, which contains the information that the model has access to plus a little more .

If you think about it that makes sense. We ideally only want to give our model the exact amount of information it needs so we don't overload its context window.

We therefore need to make a tool call for our agent to be able to access that information.

## Accessing Context

```python
from langchain.tools import tool, ToolRuntime

@tool
def get_favourite_colour(runtime: ToolRuntime) -> str:           # We are accessing from the toolRuntime!
    """Get the favourite colour of the user"""
    return runtime.context.favourite_colour

@tool
def get_least_favourite_colour(runtime: ToolRuntime) -> str:
    """Get the least favourite colour of the user"""
    return runtime.context.least_favourite_colour
```

```python
agent = create_agent(
    model=model,
    tools=[get_favourite_colour, get_least_favourite_colour],
    context_schema=ColourContext
)
```

```python
response = agent.invoke(
    {"messages": [HumanMessage(content="What is my favourite colour?")]},
    context=ColourContext()
)

pprint(response)
```

```python
print(clean_response(response))
```

```python
response = agent.invoke(
    {"messages": [HumanMessage(content="What is my favourite colour?")]},
    context=ColourContext(favourite_colour="green")
)

pprint(response)
```

```python
print(clean_response(response))
```

However, this context is immutable, meaning our agent can't actually update or change it itself!!

That could be pretty limiting.

Maybe my preferred language or company role is something the agent learns itself during our conversation and it wants to update its memory with that info!!

Luckily, we can do that using the agent's state. 

=> If we'd like this state to keep track of anything else, we just add those value to our custom state.
