### 1. <span style="font-weight:bold">Model Context Protocol (MCP): </span>
An open protocol that standardizes how your LLM applications connect to and work with your tools and data sources (defenition by Anthropic !!)

The best analogie to understand <span style="font-weight:bold">MCP</span> is USB! an USB allows you to connect a mic for example to your laptop just by plugging it in ! Just like USB-C standardized charging/data for all devices, MCP standardizes how AI agents connect to tools/data.

Before = Messy Adapters Everywhere

Your AI Agent → [OpenAI Adapter] → Tool A  
Your AI Agent → [Custom Code] → Tool B    
Your AI Agent → [Different Adapter] → Tool C       
Problem: Every tool needed custom integration. Changing models/agents meant rewriting connections.

How MCP Changed It:

After = One Universal Port                  
Your AI Agent → [MCP Standard Port] → Tool A         
Your AI Agent → [MCP Standard Port] → Tool B        
Your AI Agent → [MCP Standard Port] → Tool C       
Solution: Tools speak MCP language. Agents speak MCP language. They just connect.

What MCP Actually Does:   
Standard Language: Defines how agents say "Give me data" and tools say "Here's the data"    
Plug & Play: Add new tools without rewriting your agent     
Security: Controlled access to tools/data     
Portability: Switch AI models easily (Claude ↔ GPT ↔ Gemini)     




Creating tools and providing context to different model providers used to look a lot like this:     
A never ending web of API calls and databases to connect to your agent for every application you tried to build 

Anthropic were encountring this issue firsthand, which is why they suggested a universtal model context protocol for model providers and tool builders to use.

In this standard, the MCP host hosts an MCP client, which communicates with the MCP server.  
In our case, we can think of the MCP host as the AI application or the Agent.   
Thses MCP servers can expose tools, resources, like read only data or prompt templates to the client.   
Basically everything our agent would ever need.   

Now a question should be asked !! A vendor agnostic framework for connecting models to tools and context. It sounds a lot like they could have just used LangChain. 
While it's true that we don't need MCP to easily swap out the model in our agent, MCP still has some advantages when building with LangChain.      

Once we've built an MCP server with our tools and context, it's very easy to share with other projects or developers. Streamlining future agent builds.

There's a huge open source community of MCP servers that other people have built. Which we can easily insert into our agent and other types of AI applications are compatible with MCP servers, like your favorite chatbots or IDEs.

```python
import sys
import asyncio

# Fix for Windows issues in Jupyter notebooks
if sys.platform == "win32":
    # 1. Use ProactorEventLoop for subprocess support
    if not isinstance(asyncio.get_event_loop_policy(), asyncio.WindowsProactorEventLoopPolicy):
        asyncio.set_event_loop_policy(asyncio.WindowsProactorEventLoopPolicy())
    
    # 2. Redirect stderr to avoid fileno() error when launching MCP servers
    if "ipykernel" in sys.modules:
        sys.stderr = sys.__stderr__
```

## Local MCP server

```python
from langchain_mcp_adapters.client import MultiServerMCPClient

client = MultiServerMCPClient(
    {
        "local_server": {
                "transport": "stdio",
                "command": "python",
                "args": ["resources/2.1_mcp_server.py"],
            }
    }
)
```

```python
# get tools
tools = await client.get_tools()

# get resources
resources = await client.get_resources("local_server")

# get prompts
prompt = await client.get_prompt("local_server", "prompt")
prompt = prompt[0].content
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI
import os

# Get your key from environment
google_key = os.getenv("GOOGLE_API_KEY")
print(f"Key exists: {bool(google_key)}")

# Create the model
model = ChatGoogleGenerativeAI(
    model="models/gemini-2.5-flash",  # Use flash for free tier
    google_api_key=google_key
)
```

```python
from langchain.agents import create_agent

agent = create_agent(
    #model="gpt-5-nano",
    model= model,
    tools=tools,
    system_prompt=prompt
)
```

```python
from langchain.messages import HumanMessage

config = {"configurable": {"thread_id": "1"}}

response = await agent.ainvoke(
    {"messages": [HumanMessage(content="Tell me about the langchain-mcp-adapters library")]},
    config=config
)
```

```python
from pprint import pprint

pprint(response)
```

```python
def clean_response(response):
    """Extract clean text from agent response"""
    last_msg = response['messages'][-1]
    
    if isinstance(last_msg.content, list):
        # Handle Gemini's list format
        for item in last_msg.content:
            if isinstance(item, dict) and item.get('type') == 'text':
                return item['text']
    else:
        return last_msg.content
    
    return "No text content found"

print(clean_response(response))
```

## Online MCP

```python
client = MultiServerMCPClient(
    {
        "time": {
            "transport": "stdio",
            "command": "uvx",
            "args": [
                "mcp-server-time",
                "--local-timezone=America/New_York"
            ]
        }
    }
)

tools = await client.get_tools()
```

```python
agent = create_agent(
    model= model,
    tools=tools,
)
```

```python
question = HumanMessage(content="What time is it?")

response = await agent.ainvoke(
    {"messages": [question]}
)

pprint(response)
```

```python
print(clean_response(response))
```
