In this chapter we will learn how to add supervision to our agent with a human in the loop.

But where should we put a human in the loop ?    

After all the goal of building agent is to automate !! 
So when, or if we include a human, is as important decision as the tools we give our agent or the models we choose.        

We can boil down the many use cases of a human in the loop into three main categorizes:  

    1.Approving sensitive actions
    2.Adding missing context
    3.Debugging our agent

We're going to focus on that first use case in this video : Approving sensitive actions, the techniques we'll see here are the exact same we'd use for the two other types!

Now let's start ! 

Let's say we have an email inbox agent who can read my inbox and summarize emails for me !              
nothing risky about that.                

It can also draft and send emails on my behalf, now maybe that's an action I'd like to have the final say in !!!!                      

```python
from dotenv import load_dotenv

load_dotenv()
```

```python
from langchain.tools import tool, ToolRuntime

@tool
def read_email(runtime: ToolRuntime) -> str:
    """Read an email from the given address."""
    # take email from state
    return runtime.state["email"]

@tool
def send_email(body: str) -> str:
    """Send an email to the given address with the given subject and body."""
    # fake email sending
    return f"Email sent"
```

```python
from langchain.agents import create_agent, AgentState
from langgraph.checkpoint.memory import InMemorySaver
from langchain.agents.middleware import HumanInTheLoopMiddleware

class EmailState(AgentState):
    email: str

agent = create_agent(
    model="gpt-5-nano",
    tools=[read_email, send_email],
    state_schema=EmailState,
    checkpointer=InMemorySaver(),
    middleware=[
        HumanInTheLoopMiddleware( #Now when we initialize our agent we add this humanintheloop function !
            interrupt_on={
                "read_email": False,
                "send_email": True,
            }, #True means we require approval, False mean we dont require approval and it's going to be auto approved
            description_prefix="Tool execution requires approval",
        ),
    ],
)
```

```python
from langchain.messages import HumanMessage

config = {"configurable": {"thread_id": "1"}}

response = agent.invoke(
    {
        "messages": [HumanMessage(content="Please read my email and send a response immediately. Send the reply now in the same thread.")],
        "email": "Hi Seán, I'm going to be late for our meeting tomorrow. Can we reschedule? Best, John."
    },
    config=config
)
```

```python
from pprint import pprint

pprint(response)
```

```python
print(response['__interrupt__'])
```

```python
# Access just the 'body' argument from the tool call
print(response['__interrupt__'][0].value['action_requests'][0]['args']['body'])
```

## Approve

```python
from langgraph.types import Command

response = agent.invoke(
    Command( 
        resume={"decisions": [{"type": "approve"}]}
    ), 
    config=config # Same thread ID to resume the paused conversation
)

pprint(response)
```

## Reject

```python
response = agent.invoke(
    Command(        
        resume={
            "decisions": [
                {
                    "type": "reject",
                    # An explanation of why the request was rejected
                    "message": "No please sign off - Your merciful leader, Seán."
                }
            ]
        }
    ), 
    config=config # Same thread ID to resume the paused conversation
    )   

pprint(response)
```

```python
print(response['__interrupt__'][0].value['action_requests'][0]['args']['body'])
```

## Edit

```python
response = agent.invoke(
    Command(        
        resume={
            "decisions": [
                {
                    "type": "edit",
                    # Edited action with tool name and args
                    "edited_action": {
                        # Tool name to call.
                        # Will usually be the same as the original action.
                        "name": "send_email",
                        # Arguments to pass to the tool.
                        "args": {"body": "This is the last straw, you're fired!"},
                    }
                }
            ]
        }
    ), 
    config=config # Same thread ID to resume the paused conversation
    )   

pprint(response)
```
