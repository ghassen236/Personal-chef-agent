What separates an agent from a standard chatbot is its ability to take actions, percieve the output of those actions, and react accordingly. 

React Agents use this pattern scince most of the industry has coalesced (fusionner) around this method, we'll call react Agents Agents 

The action that an agent can take are defined by the tools that we provide it . 

Tools can allow our Agent to access data, execute tasks, even call other Agents !, transforming it from a passive language model to the coordinator of a much more capable system

We define a tool in langChain the same way we define any other function, with a few extra things to keep in mind. (like the tool decorator that well see in the code below ! @tool)

## Tool definition

```python
from dotenv import load_dotenv

load_dotenv()
```

The @tool decorator IS from LangChain (from langchain.tools import tool). It's LangChain-specific and tells LangChain: "Hey, treat this function as a tool my agent can use."

There are Other Python decorators (but they are DIFFERENT)




##### Decorator: A function that modifies another function
##### Simple definition: A decorator wraps a function to add behavior without changing its code.



```python
from langchain.tools import tool

@tool #It turns it from a function to a tool that the Agent can use.
def square_root(x: float) -> float:
    """Calculate the square root of a number""" #description for the Agent !!
    return x ** 0.5

```

##### We need to make sure that our tool's name and description are detailed enough for our agent to understand what the tool is and what it actually does. By default, the tool's name comes from the function's name, but we can override it with a custom name before the tool decorator like the example bellow ! The same for the description ! We can also override the description


```python
@tool("square_root")
def tool1(x: float) -> float:
    """Calculate the square root of a number"""
    return x ** 0.5
```

```python
@tool("square_root", description="Calculate the square root of a number")
def tool1(x: float) -> float:
    return x ** 0.5
```

```python
tool1.invoke({"x": 467})
```

## Adding to agents

```python
from langchain_groq import ChatGroq
model = ChatGroq(model="llama-3.3-70b-versatile")
```

```python
from langchain.agents import create_agent

agent = create_agent(
    #model="gpt-5-nano",
    model= model,
    tools=[tool1],
    system_prompt="You are an arithmetic wizard. Use your tools to calculate the square root and square of any number."
)
```

```python
from langchain.messages import HumanMessage

question = HumanMessage(content="What is the square root of 467?")

response = agent.invoke(
    {"messages": [question]}
)

print(response['messages'][-1].content)
```

```python
from pprint import pprint

pprint(response['messages'])
```

```python
print(response["messages"][1].tool_calls)
```
