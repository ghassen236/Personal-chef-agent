````python
from dotenv import load_dotenv

load_dotenv()
````

## No memory

````python
from langchain_groq import ChatGroq
model = ChatGroq(model="llama-3.3-70b-versatile")
````

````python
from langchain.agents import create_agent


agent = create_agent(
    #"gpt-5-nano"
    model= model
)
````

````python
from langchain.messages import HumanMessage

question = HumanMessage(content="Hello my name is Seán and my favourite colour is green")

response = agent.invoke(
    {"messages": [question]} 
)
````

````python
from pprint import pprint

pprint(response)
````

````python
print(response['messages'][-1].content)
````

````python
question = HumanMessage(content="What's my favourite colour?")

response = agent.invoke(
    {"messages": [question]} 
)

pprint(response)
````

````python
print(response['messages'][-1].content)
````

#### It forgotten all about our previous interaction ! 

So what's happening here with our LangChain agent, we're tracking messages in something called the state, which you can think of aas the memory of our agent.

The problem is the state isn't being saved from one run to another, so in effect, our agent's memory is being wiped.

What we need to do is somehow save our states that the agent can remember previous messages.

We do that by using what's called checkpointer. This saves a snapshot of the state at the end of each run, and then groups together runs with the same thread id.

The chekpointer we use is imported from the LangGraph Library, it's called InMemorySaver.

## Memory

````python
from langgraph.checkpoint.memory import InMemorySaver  


agent = create_agent(
    #"gpt-5-nano",
    model= model,
    checkpointer=InMemorySaver(),  
)
````

````python
from langchain.messages import HumanMessage

question = HumanMessage(content="Hello my name is LMhaf and my favourite colour is green")
config = {"configurable": {"thread_id": "1"}}

response = agent.invoke(
    {"messages": [question]},
    config,  
)
````

````python
pprint(response)
````

````python
print(response['messages'][-1].content)
````

````python
question = HumanMessage(content="What's my favourite colour?")

response = agent.invoke(
    {"messages": [question]},
    config,  
)

pprint(response)
````

````python
print(response['messages'][-1].content)
````

### 64-bit vs 32-bit: The Highway Analogy
##### Think of it as highway lanes for data:

| Aspect                | 32-bit                         | 64-bit                                  |
|----------------------|--------------------------------|------------------------------------------|
| Lanes (Data Path)     | 32 lanes                       | 64 lanes                                 |
| Max Memory Addresses  | ~4 billion                     | ~18 quintillion                          |
| Speed                 | Slower                         | Faster                                   |
| Modern Use            | Old phones, legacy apps        | Everything today (post-2005)             |
| File/App Labels       | x86, i386, 32-bit              | x64, amd64, 64-bit                       |
| Typical RAM Support   | Up to 4 GB                     | 16 GB+ (theoretically 16 exabytes)       |
| Python Installers     | python-3.13.2.exe (32-bit)     | python-3.13.2-amd64.exe (64-bit)         |


"Bit" means binary digit (0 or 1). It's the smallest unit of data in computing.




"32-bit" = A processor with 32 data wires that can only count to ~4 billion and handle max 4GB RAM.

"64-bit" = 64 wires, counts to 18 quintillion, handles virtually unlimited RAM.



"How strong is your processor's data highway?"

32-bit = 2-lane country road (slower, limited capacity)

64-bit = 8-lane superhighway (faster, massive capacity)

config = {"configurable": {"thread_id": "chef_session_1"}}
```

Breaking it down:

- `"configurable"` — this is LangGraph's standard key for runtime settings, always required, never changes
- `"thread_id"` — your conversation label, you choose the value
- `"chef_session_1"` — just a name you invented, could be anything

A mental model:
```
InMemorySaver (the filing cabinet)
│
├── thread_id: "chef_session_1"  ← your current conversation
│     ├── Turn 1: [image + "what can I cook?"]
│     ├── Turn 1 response: ["frittata, soup, scramble..."]
│     └── Turn 2: ["give me the frittata recipe"]
│
├── thread_id: "chef_session_2"  ← a different user/conversation
│     └── Turn 1: ["I have pasta and garlic..."]
│
└── thread_id: "chef_session_3"  ← yet another conversation
