## Customising your Model

```python
# model = init_chat_model(
#     model="gpt-5-nano",
#     # Kwargs passed to the model:
#     temperature=1.0
# )

 
model = ChatGroq(
    model = "llama-3.3-70b-versatile",
    # Kwargs passed to the model:
    temperature=1
)

response = model.invoke("What's the capital of the Moon?")
print(response.content)
```

```python
response
```

### We can customize our model outta the box by adjusting a few parameters. Like:

#### temperature:
    .that controles the randomeness of the model's output a higher temperature number makes responses       more creative, whereas a lower one is going to make them more deterministic.
#### Max tokens:
    .which is going to limit the total number of tokens in the response effectively controlling how   long the output can be.
#### Timeout: 
    .which is going to be the max time to wait for a response from the model before cancelling their request.
#### Max_retries:
    .the max amount of times to retry a request if that request fails.

## Initialising and invoking an agent

```python
from langchain.agents import create_agent

agent = create_agent(model=model)
```

```python
# agent = create_agent(model="claude-sonnet-4-5")
# We can specify the name of the model 
```

```python
# agent = create_agent("gpt-5-nano")
```

```python
from langchain.messages import HumanMessage

response = agent.invoke(
    {"messages": [HumanMessage(content="What's the capital of the Moon?")]}
)
```

```python
from pprint import pprint

pprint(response)
```

```python
print(response['messages'][-1].content)
```

```python
from langchain.messages import AIMessage

response = agent.invoke(
    {"messages": [HumanMessage(content="What's the capital of the Moon?"),   # We ask 
    AIMessage(content="The capital of the Moon is Luna City."),              # We force this answer
    HumanMessage(content="Interesting, tell me more about Luna City")]}      # We follow up 
)

pprint(response)
```

```python
print(response['messages'][-1].content)
```

## Streaming Output

```python
for token, metadata in agent.stream(
    {"messages": [HumanMessage(content="Tell me all about Luna City, the capital of the Moon")]},
    stream_mode="messages"
):

    # token is a message chunk with token content
    # metadata contains which node produced the token
    
    if token.content:  # Check if there's actual content
        print(token.content, end="", flush=True)  # Print token
```
