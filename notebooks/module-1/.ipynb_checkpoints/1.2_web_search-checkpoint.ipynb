## Without web search

```python
from dotenv import load_dotenv

load_dotenv()
```

```python
from langchain_groq import ChatGroq
model = ChatGroq(model="llama-3.3-70b-versatile")
```

```python
from langchain.agents import create_agent

agent = create_agent(
    #model="gpt-5-nano"
    model= model 
)
```

```python
from langchain.messages import HumanMessage

question = HumanMessage(content="How up to date is your training knowledge?")

response = agent.invoke(
    {"messages": [question]}
)
```

```python
print(response['messages'][-1].content)
```

```python
from langchain.messages import HumanMessage

question = HumanMessage(content="Who is the current mayor of San Francisco?")

response = agent.invoke(
    {"messages": [question]}
)

print(response['messages'][-1].content)
```

#### Any question about an event that happened after that date, the model will get it wrong !


## Add web search tool

```python
from langchain.tools import tool
from typing import Dict, Any
from tavily import TavilyClient #Tavily is a search engine build for AIs. Unlike google which gives you web pages, Tavily: 
                                    #searches the web for our AI
                                    #Summarize the results 
                                    #Return clean, structured information that AI can understand easily
                                    #Most AIs cannot directly browse the web, they need "tools" !!

tavily_client = TavilyClient()

@tool
def web_search(query: str) -> Dict[str, Any]:   #input/output types    ;; normal functions don't need this (-> .... ) . It's optional in Python 
                                                                        #but crucial for tools because: Tells developers what the tool returns; Agent knows what to expect...

    """Search the web for information"""

    return tavily_client.search(query)

web_search.invoke("Who is the current mayor of San Francisco?")
```

##### 🤔 Does the Agent ALWAYS use tools?                    
NO! The agent decides when to use tools based on:

The question (math question → use calculator)

The tool descriptions (what each tool does)

The system prompt (your instructions)
